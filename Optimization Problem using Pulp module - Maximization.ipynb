#### Initialize The Model

```python
# Import all classes of PuLP module
from pulp import *

# Create the problem variable to contain the problem data
model = LpProblem("Production_Problem", LpMaximize)
```

#### Define Decision Variable

```python
# Creating 25 decision variable for the 25 different product


x1 = LpVariable("product 1", 0, None, LpInteger)
x2 = LpVariable("product 2", 0, None, LpInteger) 
x3 = LpVariable("product 3", 0, None, LpInteger)
x4 = LpVariable("product 4", 0, None, LpInteger)
x5 = LpVariable("product 5", 0, None, LpInteger) 

x6 = LpVariable("product 6", 0, None, LpInteger)
x7 = LpVariable("product 7", 0, None, LpInteger) 
x8 = LpVariable("product 8", 0, None, LpInteger)
x9 = LpVariable("product 9", 0, None, LpInteger)
x10 = LpVariable("product 10", 0, None, LpInteger) 


x11= LpVariable("product 11", 0, None, LpInteger)
x12= LpVariable("product 12", 0, None, LpInteger) 
x13= LpVariable("product 13", 0, None, LpInteger)
x14= LpVariable("product 14", 0, None, LpInteger)
x15= LpVariable("product 15", 0, None, LpInteger) 


x16= LpVariable("product 16", 0, None, LpInteger)
x17= LpVariable("product 17", 0, None, LpInteger) 
x18= LpVariable("product 18", 0, None, LpInteger)
x19= LpVariable("product 19", 0, None, LpInteger)
x20= LpVariable("product 20", 0, None, LpInteger) 


x21= LpVariable("product 21", 0, None, LpInteger)
x22= LpVariable("product 22", 0, None, LpInteger) 
x23= LpVariable("product 23", 0, None, LpInteger)
x24= LpVariable("product 24", 0, None, LpInteger)
x25= LpVariable("product 25", 0, None, LpInteger) 
```

#### Define Objective Function

```python
# Create maximize objective function
model += 3.8 * x1 + 2.91 * x2 + 5.73 * x3 + 11.17 * x4 + 1.57 * x5 + 5.67 * x6 + 4.72 * x7 + 3.37 * x8 + 3.39 * x9 + 7.02 * x10 + 2.24 * x11 + 3.54 * x12 + 1.95 * x13 + 4.6 * x14 + 2.52 * x15 + 3.27 * x16 + 3.66 * x17 + 1.88 * x18 + 3.71 * x19 + 3.4 * x20 + 3.32 * x21 + 1.7 * x22 + 4.19 * x23 + 2.65 * x24 + 2.38 * x25
```

#### Define Constrains

```python
# Create three constraints
model += 1.76 * x1 + 1.2 * x2 + 3 * x3+ 4.8 * x4 + 0.6 * x5 + 2.4 * x6 + 1.92 * x7 + 1.44 * x8 + 1.44 * x9 + 2.4 * x10 + 0.6 * x11 + 0.6 * x12 + 0.6 * x13 + 1.5 * x14 +0.75 * x15 + 0.24 * x16 + 0.75 * x17 + 0.3 * x18 + 0.9 * x19 + 0.6 * x20 + 0.3 * x21 + 1.2 * x22 +2 * x23 + 1.08 * x24 + 0.45 * x25 <= 175000, "Raw Material"
model += 0.4 * x1 + 0.66 * x2 + 0.24 * x3 + 0.88 * x4 + 0.22 * x5 + 1.33 * x6 + 0.88 * x7 + 0.3 * x8 + 0.48 * x9 + 6.6 * x10 + 0.34 * x11 + 0.83 * x12 + 0.45 * x13 + 0.48 * x14 + 0.66 * x15 + 0.24 * x16 + 0.9 * x17 + 0.3 * x18 + 0.9 * x19 + 0.83 * x20 + 0.3 * x21 + 0.4 * x22 + 1.14 * x23 + 0.48 * x24 + 0.45 * x25 <= 168000, "Machine Capacity"
model += 0.46 * x1 + 0.75 * x2 + 0.27 * x3 + 1 * x4 + 0.24 * x5 + 1.52 * x6 + 1 * x7 + 0.34 * x8 + 0.55 * x9 + 7.52 * x10 + 0.39 * x11 + 0.95 * x12 + 0.51 * x13 + 0.55 * x14 + 0.75 * x15 + 0.27 * x16 + 1.03 * x17 + 0.34 * x18 + 1.03 * x19 + 0.95 * x20 + 0.34 * x21 + 0.46 * x22 + 1.3 * x23 + 0.55 * x24 + 0.51 * x25 <= 192000, "Labour Capacity"
#Maximum average demand
model += x1 <= 4000, "product1"
model += x2 <= 15000, "product2"
model += x3 <= 2100, "product3"
model += x4 <= 1000, "product4"
model += x5 <= 20000, "product5"

model += x6 <= 1200,"product6"
model += x7 <= 5000, "product7"
model += x8 <= 12000, "product8"
model += x9 <= 6200, "product9"
model += x10 <= 3000, "product10"

model += x11 <= 14000, "product11"
model += x12<= 8200, "product12"
model += x13<= 16000, "product13"
model += x14<= 5000, "product14"
model += x15<= 12000, "product15"

model += x16<= 22000, "product16"
model += x17<= 6500, "product17"
model += x18<= 18000, "product18"
model += x19<= 4000, "product19"
model += x20<= 7500, "product20"

model += x21<= 9000, "product21"
model += x22<= 7500, "product22"
model += x23<= 3500, "product23"
model += x24<= 4500, "product24"
model += x25<= 10000, "product25"
```

```python
# The problem is solved using PuLP's choice of Solver
model.solve()

# Each of the variables is printed with it's resolved optimum value
for v in model.variables():
    print(v.name, "=", v.varValue)
print("Max Profit = ",value(model.objective ))
```
